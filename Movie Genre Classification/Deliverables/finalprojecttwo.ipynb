```python
def labelTag(listTag,key):
    listvalue = []
    for i in listTag:
        if key in i:#Check if the key is inside the list
            listvalue.append(1)
        else:
            listvalue.append(0)
    return listvalue
```

# Action

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(matrix.toarray(),moviesTruth['actiontruth'],test_size=0.2,random_state=4)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(x_train, y_train)
```

```python
conf = 'Confusion matrix: '
prec = 'Precision score: '
rec = 'Recall score: '
f1s = 'F1 score: '
```

# SVM

```python
from sklearn import metrics


# STEP 3: make predictions on the testing set
y_pred = clf.predict(x_test)


# compare actual response values (y_test) with predicted response values (y_pred) 

# you need to set parameter "average" because this is multi-class classfier
print('SVM Train Metrics')
print('\n' + conf)
print(metrics.confusion_matrix(y_test,y_pred))
print('\n' + prec)
print(metrics.precision_score(y_test,y_pred, average='weighted'))   # Parameter "average" is requred if not a binary model
print('\n' + rec)
print(metrics.recall_score(y_test,y_pred, average='weighted'))     # Parameter "average" is requred if not a binary model
print('\n'+ f1s)
print(metrics.f1_score(y_test,y_pred, average='weighted'))         # Parameter "average" is requred if not a binary model
```
